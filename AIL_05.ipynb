```python
a1 = -0.1
a2 = -1.5
a3 = 3
a4 = 8

x2 = np.linspace(start = -8, stop = 8, num=10000).reshape(-1,1)
y2 = np.array([(a1*i**2 + a2*np.sin(i) + a3*np.log10(i**2+1)) for i in x2])
```

```python
plt.plot(x2, y2, 'r')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```
